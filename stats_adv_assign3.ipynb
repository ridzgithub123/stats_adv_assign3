```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation statistics is a branch of statistics concerned with estimating population parameters based on sample data. 
Population parameters are numerical characteristics of a population, such as the population mean, population proportion,population variance, etc.
Estimation statistics aims to make inferences about these parameters using information obtained from a sample drawn from the population.

Point Estimate:
A point estimate is a single value that serves as the best guess or approximation of a population parameter based on sample data. 
It is obtained by calculating a sample statistic, such as the sample mean or sample proportion, 
and using it as an estimate of the corresponding population parameter. Point estimates provide a single value as an estimate of
the population parameter but do not provide information about the uncertainty associated with the estimate.

Interval Estimate:
An interval estimate, also known as a confidence interval, is a range of values that is likely to containthe true value of the
population parameter with a certain level of confidence. It is obtained by calculating a range of values around the point estimate, 
called the confidence interval, which is constructed based on the variability of the sample data and the desired level of confidence.
Interval estimates provide a range of values within which the true population parameter is expected to lie, along with a measure 
of the uncertainty associated with the estimate.

Example:
Suppose we want to estimate the average height of adult males in a city.
We take a random sample of 100 adult males from the population and calculate the sample mean height, which turns out to be 175 centimeters.

Point Estimate: The sample mean height of 175 centimeters serves as the point estimate of the population mean height.

Interval Estimate: We construct a 95% confidence interval around the point estimate, which might be [170, 180] centimeters.
This interval estimate indicates that we are 95% confident that the true population mean height falls within the range of 170 to 180 centimeters.
```

```python
# Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data. 
It involves formulating a hypothesis about a population parameter, collecting sample data, and 
then using statistical tests to determine whether the evidence from the sample provides enough support to either accept or reject the hypothesis.

The two primary types of hypotheses involved in hypothesis testing are:

Null Hypothesis (H0): This hypothesis represents the status quo or a default assumption.
It states that there is no significant difference, effect, or relationship between variables in the population.

Alternative Hypothesis (H1 or Ha): This hypothesis represents the research question or the hypothesis that contradicts the null hypothesis.
It states that there is a significant difference, effect, or relationship between variables in the population.

Hypothesis testing is used to assess the validity of claims, evaluate the effectiveness of interventions or treatments, compare groups or populations,
and make decisions based on available evidence. 

It serves several important purposes:

Evidence-Based Decision Making: Hypothesis testing provides a systematic framework for making decisions based on empirical evidence
rather than intuition or guesswork. By testing hypotheses, researchers and decision-makers can evaluate the effectiveness of 
interventions, policies, or strategies and make informed decisions.

Scientific Inquiry: Hypothesis testing is an essential tool in scientific research for testing theories,
exploring relationships between variables, and advancing knowledge in various fields.
It allows researchers to validate or refute hypotheses, leading to the accumulation of scientific knowledge and the refinement of theories.

Quality Control and Process Improvement: In industries such as manufacturing and quality control,
hypothesis testing is used to evaluate processes, detect defects or deviations from standards, 
and implement corrective actions to improve quality and efficiency.

Inferential Statistics: Hypothesis testing is a fundamental component of inferential statistics, 
which involves making inferences about populations based on sample data. 
By testing hypotheses about population parameters, researchers can draw conclusions about the population as a whole.

Legal and Regulatory Compliance: Hypothesis testing is often used in legal and 
regulatory contexts to assess compliance with laws, regulations, and standards. 

For example, it may be used to determine whether a company's practices comply with anti-discrimination laws or whether a
medical treatment meets regulatory requirements.
```

```python
#Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
#given a sample from each population.

from scipy import stats

def hypothesis_test(sample1, sample2, alpha=0.05):
    
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    
    reject_null = p_value < alpha
    
    return t_statistic, p_value, reject_null


sample1 = [82, 85, 78, 80, 86, 88, 92, 95, 91, 84]
sample2 = [75, 79, 80, 73, 82, 81, 76, 78, 80, 85]

t_stat, p_val, reject_null = hypothesis_test(sample1, sample2)

print("t-statistic:", t_stat)
print("p-value:", p_val)
print("Reject null hypothesis?", reject_null)
```

```python
Q8. Define p-value and explain its significance in hypothesis testing.

The p-value, or probability value, is a measure of the evidence against a null hypothesis in hypothesis testing.
It quantifies the probability of observing the test statistic (or a more extreme value) if the null hypothesis is true.
In other words, the p-value indicates the likelihood of obtaining the observed result, or a result more extreme,
under the assumption that the null hypothesis is correct.

Significance of the p-value in Hypothesis Testing:

Decision Making: The p-value serves as a basis for decision-making in hypothesis testing.
If the p-value is less than or equal to a predetermined significance level (usually denoted as α), typically 0.05, 
the null hypothesis is rejected in favor of the alternative hypothesis. Conversely, if the p-value is greater than the significance level
the null hypothesis is not rejected.

Interpretation: A small p-value indicates strong evidence against the null hypothesis, 
suggesting that the observed result is unlikely to have occurred by chance alone if the null hypothesis were true.
On the other hand, a large p-value suggests weak evidence against the null hypothesis, indicating that the observed result 
is plausible under the null hypothesis.

Error Control: Controlling the significance level (alpha) allows researchers to manage the risk of making Type I errors 
(false positives) and Type II errors (false negatives) in hypothesis testing. By choosing a lower significance level, 
such as 0.01 instead of 0.05, researchers can reduce the probability of incorrectly rejecting the null hypothesis 
when it is actually true (Type I error).

Inference: The p-value provides information about the strength of the evidence against the null hypothesis 
and helps researchers draw conclusions about the population based on sample data. 
A smaller p-value indicates stronger evidence for the alternative hypothesis, supporting the researcher's hypothesis or research question.

Replication and Generalization: When the p-value is small, indicating strong evidence against the null hypothesis,
the findings are more likely to be replicable and generalizable to the population. Conversely, 
ha large p-value suggests that the observed result is consistent with random variation and may not be reliable or generalizable.
```

```python
#Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
#parameter set to 10.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

df = 10

x = np.linspace(t.ppf(0.001, df), t.ppf(0.999, df), 1000)

pdf = t.pdf(x, df)

plt.plot(x, pdf, label='t-distribution (df=10)')
plt.title("Student's t-distribution")
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
#random samples of equal size and a null hypothesis that the population means are equal.

from scipy.stats import ttest_ind

def two_sample_t_test(sample1, sample2, alpha=0.05):

    t_statistic, p_value = ttest_ind(sample1, sample2)
    

    reject_null = p_value < alpha
    
    return t_statistic, p_value, reject_null

sample1 = [82, 85, 78, 80, 86, 88, 92, 95, 91, 84]
sample2 = [75, 79, 80, 73, 82, 81, 76, 78, 80, 85]

t_stat, p_val, reject_null = two_sample_t_test(sample1, sample2)

print("t-statistic:", t_stat)
print("p-value:", p_val)
print("Reject null hypothesis?", reject_null)
```

```python
Q11: What is Student’s t distribution? When to use the t-Distribution.

Student's t-distribution, often referred to simply as the t-distribution, is a probability distribution that 
arises when estimating the mean of a normally distributed population when the sample size is small 
and/or the population standard deviation is unknown.

Characteristics of the t-Distribution:

The t-distribution is similar in shape to the standard normal distribution (i.e., bell-shaped), but with heavier tails.
The shape of the t-distribution depends on the degrees of freedom parameter (df), which is related to the sample size. 
As df increases, the t-distribution approaches the standard normal distribution.
The mean of the t-distribution is 0, and its variance depends on the degrees of freedom parameter.
When to Use the t-Distribution:
The t-distribution is commonly used in the following scenarios:

Small Sample Sizes: When the sample size is small (typically n < 30), 
the sample standard deviation tends to underestimate the population standard deviation. 
In such cases, using the t-distribution allows for more accurate estimation of confidence intervals and hypothesis tests for the population mean.

Unknown Population Standard Deviation: When the population standard deviation is unknown, 
and the sample size is small, the t-distribution is used to estimate the population mean.
In contrast, when the population standard deviation is known and the sample size is large, 
the z-distribution (standard normal distribution) is used.

Estimating Confidence Intervals and Conducting Hypothesis Tests: The t-distribution is used to calculate confidence intervals 
and conduct hypothesis tests for the population mean, especially when dealing with small sample sizes or unknown population standard deviations.

Comparing Means of Two Independent Samples: When comparing the means of two independent samples (e.g., in a two-sample t-test),
the t-distribution is used to assess whether the difference between sample means is statistically significant.

We use t -test when we don't know the population standard deviation.
```

```python
Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.


To test the researcher's hypothesis that the new drug decreases blood pressure by 10 mmHg, we can perform a one-sample t-test. Here's the null and alternative hypothesis:

Null Hypothesis (H0): The mean decrease in blood pressure is not significantly different from 10 mmHg (i.e., μ=10).
Alternative Hypothesis (H1): The mean decrease in blood pressure is significantly less than 10 mmHg (i.e., μ<10).
We will use a significance level (α) of 0.05.

The formula to calculate the t-statistic for a one-sample t-test is:
         _
    t=  (x - μ0 )/ (s/n^1/2)
    
Where:
_
x = sample mean decrease in blood pressure
μ0 = hypothesized population mean decrease in blood pressure (in this case, 10 mmHg)
s = sample standard deviation
n = sample size
Given:
_
x = 8 mmHg

μ0 = 10 mmHg

s = 3 mmHg

n = 100
Let's calculate the t-statistic:
After putting value in formula the calc value is t≈−6.67

Now, we need to compare this t-statistic with the critical t-value from the t-distribution table with 
n−1 degrees of freedom (where n is the sample size) at a significance level of 0.05 (since this is a one-tailed test).

For n−1=99 degrees of freedom, the critical t-value is approximately -1.660 (since we are testing for μ<10 and it's a one-tailed test).

Since our calculated t-statistic (-6.67) is much smaller (i.e., more negative) than the critical t-value (-1.660), we reject the null hypothesis.

Therefore, we have sufficient evidence to conclude that the new drug significantly decreases blood pressure by more than 10 mmHg at a significance level of 0.05.





```
